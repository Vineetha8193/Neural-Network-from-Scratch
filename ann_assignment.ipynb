```python
import numpy as np

def gradient_descent(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate):
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    return W1, b1, W2, b2

# Example usage:
np.random.seed(1)
W1 = np.random.randn(4, 3) * 0.01  # Randomly initialized weights
b1 = np.zeros((4, 1))
W2 = np.random.randn(1, 4) * 0.01  # Randomly initialized weights
b2 = np.zeros((1, 1))
dW1 = np.random.randn(4, 3) * 0.01  # Example gradients
db1 = np.zeros((4, 1))
dW2 = np.random.randn(1, 4) * 0.01
db2 = np.zeros((1, 1))
learning_rate = 0.01

# Perform gradient descent update
W1, b1, W2, b2 = gradient_descent(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate)

print("Updated parameters after gradient descent:")
print(f"W1:\n{W1}")
print(f"b1:\n{b1}")
print(f"W2:\n{W2}")
print(f"b2:\n{b2}")
```

```python
import numpy as np

# Sigmoid activation function
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

# Sigmoid derivative
def sigmoid_derivative(A):
    return A * (1 - A)

# Forward propagation
def forward_propagation(X, W1, b1, W2, b2):
    Z1 = np.dot(W1, X) + b1
    A1 = sigmoid(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = sigmoid(Z2)
    return Z1, A1, Z2, A2

# Backpropagation
def backward_propagation(X, Y, Z1, A1, Z2, A2, W2):
    m = X.shape[1]
    dZ2 = A2 - Y
    dW2 = np.dot(dZ2, A1.T) / m
    db2 = np.sum(dZ2, axis=1, keepdims=True) / m
    dZ1 = np.dot(W2.T, dZ2) * sigmoid_derivative(A1)
    dW1 = np.dot(dZ1, X.T) / m
    db1 = np.sum(dZ1, axis=1, keepdims=True) / m
    return dW1, db1, dW2, db2

# Training function
def train(X, Y, hidden_size, epochs=5000, learning_rate=0.1):
    input_size = X.shape[0]
    output_size = Y.shape[0]

    # Xavier Initialization for better convergence
    W1 = np.random.randn(hidden_size, input_size) * np.sqrt(1. / input_size)
    b1 = np.zeros((hidden_size, 1))
    W2 = np.random.randn(output_size, hidden_size) * np.sqrt(1. / hidden_size)
    b2 = np.zeros((output_size, 1))

    # Training loop
    for epoch in range(epochs):
        # Forward propagation
        Z1, A1, Z2, A2 = forward_propagation(X, W1, b1, W2, b2)

        # Compute loss (Binary Cross-Entropy)
        loss = -np.mean(Y * np.log(A2) + (1 - Y) * np.log(1 - A2))

        # Backpropagation
        dW1, db1, dW2, db2 = backward_propagation(X, Y, Z1, A1, Z2, A2, W2)

        # Gradient descent update
        W1 -= learning_rate * dW1
        b1 -= learning_rate * db1
        W2 -= learning_rate * dW2
        b2 -= learning_rate * db2

        # Print loss at intervals
        if epoch % 500 == 0:
            print(f"Epoch {epoch}, Loss: {loss}")

    return W1, b1, W2, b2

# Example dataset (3 features, 5 samples)
np.random.seed(1)
X = np.random.randn(3, 5)
Y = np.array([[1, 0, 1, 0, 1]])  # Ensuring both 0s and 1s in labels

# Train the model
W1, b1, W2, b2 = train(X, Y, hidden_size=4, epochs=5000, learning_rate=0.1)

# Make predictions
_, _, _, A2 = forward_propagation(X, W1, b1, W2, b2)
predictions = (A2 > 0.5).astype(int)

print("\nFinal Predictions:")
print(predictions)
```
